```python
def search_vocab_in_file(filepath):
    vocab_list = []
    with open(filepath, mode='r', encoding='utf-8') as f:
        for line in f:
            if '__Vocabulary__' in line:
                f.readline()
                for line in f:
                    if line == '\n':
                        break
                    elif re.search('<ruby>.+</ruby>', line):
                        rr = re.search('<ruby>.+</ruby>', line) # regex result
                        sub_s = line[rr.start()+6:rr.end()-7]
                        vocab_list.append(sub_s)
                        print('\t', sub_s)
                        # print('\t', repr(line))
    
    return vocab_list
```

```python
def search_vocab_in_dir(vocab_list:list, path:str):
    if os.path.isfile(path) and re.search('\.md', path):
        print(path)
        vocab_list.extend(search_vocab_in_file(path))
    elif os.path.isdir(path):
        file_list = os.listdir(path)
        for file in file_list:
            filepath = f'{path}/{file}'
            search_vocab_in_dir(vocab_list, filepath)
```

```python
vocab_list = []
search_vocab_in_dir(vocab_list, '.')
```

```python
def base_vocab(ruby_code:str):
    base_vocab = re.sub(r'<rt>([^<^>]+)?</rt>', '', ruby_code)

    return base_vocab
```

```python
vocab_map_list = []
for vocab in vocab_list:
    vocab_map_list.append([base_vocab(vocab), vocab])
    print(base_vocab(vocab), '-'*4, vocab)
```
